```python
EXCEPTIONS = {2:[], 3:[], 4:[4], 5:[4], 6:[3], 
              7: [3, 5, 6, 7], 
              8: [5, 7, 8],
              9: [3, 4, 8],    
             10: [3, 4, 9],
             11: [3, 4, 9, 11],
             12: [6, 8, 11, 12],
             13: [4, 5, 7, 9, 10, 11, 12, 13],
             14: [3, 4, 5, 6, 7, 11, 12, 14],
             15: [5, 6, 7, 9, 10, 11, 13, 14],
             16: [4, 7, 12, 13, 14 ,16],
             17: [4, 9, 10, 13, 15, 16, 17],
             18: [3, 5, 8, 9, 11, 12, 13, 14, 15],
             19: [3, 4, 6, 8, 15, 16],
             20: [5, 9, 11, 14, 17, 18, 19],
             21: [3, 5, 8, 9, 11, 12, 18],
             22: [3, 4, 6, 8, 9, 13, 19, 20, 21],
             23: [3, 4, 5, 8, 11, 13, 14, 15, 17 ,18 ,19, 20],
             24: [6, 11, 12, 14, 16, 20 , 21, 23, 24],
             }                    
```

```python
def positions_init(nb_tables):
    """ Renvoie une liste des positions initiales en fonction du nb de tables """
    exceptions = EXCEPTIONS[nb_tables]
    positions_tables = {}
    for paire in range(1, nb_tables+1):
        positions_tables[paire] = [paire, 2*nb_tables+1 - paire] if paire in exceptions else [2*nb_tables+1 - paire, paire]
    return positions_tables
```

```python
def positions_tables2paires(positions_tables):
    """ Renvoie une liste des positions des paires en fonction des positions des tables """
    positions_paires = {}
    for (table, paires) in positions_tables.items():
        NS, EO = paires
        positions_paires[NS] = (table, 'NS')
        positions_paires[EO] = (table, 'EO')
    return positions_paires
```

```python
def positions_paires2tables(positions_paires):
    """ Renvoie une liste des positions des tables en fonction des positions des paires """
    positions_tables = {i+1:[None, None] for i in range(len(positions_paires)//2)}
    for (paire, position) in positions_paires.items():
        table, orientation = position
        if orientation == 'NS':
            positions_tables[table][0] = paire
        else:
            positions_tables[table][1] = paire
    return positions_tables
```

```python
def affiche_positions_paires(positions_paires):
    for paire in sorted(positions_paires):
        table, orientation = positions_paires[paire]
        print(f"paire n°{paire} : table n°{table} - {orientation}")
```

```python
def affiche_positions_tables(positions_tables):
    for table in sorted(positions_tables):
        NS, EO = positions_tables[table]
        print(f"table n°{table} : NS={NS}, EO={EO}")
```

```python
def mouvement(positions_tables):
    """ Renvoie la nouvelle carte des positions après un changement de positions """
    positions_paires = positions_tables2paires(positions_tables)
    nb_tables = len(positions_tables)
    nb_paires = len(positions_paires)
    pos_dernier = positions_paires[nb_paires-1]
    for paire in range(nb_paires-1, 1, -1):
        positions_paires[paire] = positions_paires[paire-1]
    positions_paires[1] = pos_dernier
    return positions_paires2tables(positions_paires)
```

```python
with open(f"barometre_{nb_tables}_tables.csv", "w") as f:
    entete = "tour"
    for i in range(2*nb_tables):
        entete += f",table de paire {i+1},orientation de paire {i+1}"
    entete += '\n'
    f.write(entete)
    
    positions_tables = positions_init(nb_tables)
    positions_paires = positions_tables2paires(positions_tables)
    pos_init_dernier = positions_paires[len(positions_paires)-1]

    tour = 1
    ligne = str(tour)
    for paire in sorted(positions_paires):
        ligne += ',' + str(positions_paires[paire][0]) + ',' + positions_paires[paire][1]
    ligne += '\n'
    f.write(ligne)
    
    positions_tables = mouvement(positions_tables)
    positions_paires = positions_tables2paires(positions_tables)
    
    while positions_paires[len(positions_paires)-1] != pos_init_dernier:

        ligne = str(tour+1)
        for paire in sorted(positions_paires):
            ligne += ',' + str(positions_paires[paire][0]) + ',' + positions_paires[paire][1]
        ligne += '\n'
        f.write(ligne)
        
        positions_tables = mouvement(positions_tables)
        positions_paires = positions_tables2paires(positions_tables)
        tour += 1
```
